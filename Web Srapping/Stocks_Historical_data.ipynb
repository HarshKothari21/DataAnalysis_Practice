Web Srapping/Stocks_Historical_data.ipynb

```
import requests
import pandas as pd
from bs4 import BeautifulSoup

# URL to scrape
data_url = "https://finance.yahoo.com/quote/AAPL/history/?frequency=1mo&period1=1104537600&period2=1735182270"

# Headers to mimic a browser visit
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# Perform an HTTP GET request to fetch the webpage
response = requests.get(data_url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the table containing historical data
    table = soup.find("table", class_="table yf-j5d1ld")

    if table:
        # Extract table rows
        rows = table.find_all("tr")

        # Extract headers
        headers = [header.get_text(strip=True) for header in rows[0].find_all("th")]

        # Extract data rows
        data = []
        for row in rows[1:]:
            cols = row.find_all("td")
            if len(cols) > 0:  # Ensure row has data
                data.append([col.get_text(strip=True) for col in cols])

        # Create a DataFrame
        df = pd.DataFrame(data, columns=headers)

        # Display the DataFrame
        print(df)

        # Save to CSV
        df.to_csv("AAPL_historical_data.csv", index=False)
        print("Data saved to AAPL_historical_data.csv")
    else:
        print("Historical data table not found on the page.")
else:
    print(f"Failed to fetch the webpage. Status code: {response.status_code}")
```
